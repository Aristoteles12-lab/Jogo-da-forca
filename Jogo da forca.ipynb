```python
#Psudocódigo


#1.Definir uma lista de palavras possíveis.

#2.Escolher aleatoriamente uma palavra da lista como a palavra secreta.

#3.Criar uma representação oculta da palavra secreta usando "_" para cada letra.

#4.Definir o número máximo de tentativas permitidas (por exemplo, 6).

#5.Criar um conjunto ou lista para armazenar letras já tentadas.

#6.Enquanto o número de tentativas for maior que zero e ainda houver "_" na palavra oculta, fazer:
#6.1. Mostrar ao jogador a palavra oculta atual.
#6.2. Mostrar letras já tentadas.
#6.3. Mostrar número de tentativas restantes.
#6.4. Solicitar uma letra ao jogador.
#6.5. Se a letra já foi tentada, informar o jogador e voltar ao início do loop.
#6.6. Adicionar a letra às letras já tentadas.
#6.7. Se a letra estiver na palavra secreta:
#6.7.1. Substituir os "_" correspondentes pela letra em todas as posições corretas.
#6.7.2. Informar que a letra está correta.
#6.8. Caso contrário:
#6.8.1. Reduzir o número de tentativas em 1.
#6.8.2. Informar que a letra está incorreta.

#7.Se não houver mais "_" na palavra oculta, informar que o jogador venceu.

#8.Caso contrário, informar que o jogador perdeu e revelar a palavra secreta.
```

```python
# Import
import random
from os import system, name
```

```python
# Função que desenha a forca na tela
def display_hangman(chances):

    # Lista de estágios da forca
    stages = [  # estágio 6 (final)
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # estágio 5
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                # estágio 4
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                # estágio 3
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                # estágio 2
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # estágio 1
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # estágio 0
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[chances]
```

```python
# Função do jogo
def game():

    limpa_tela()
    print("\nBem vindo ao jogo da forca!")
    print("Adivinhe a palavra abaixo:\n")
    
    # Lista de palavras para o jogo
    palavras = ['banana', 'abacate', 'uva', 'morango', 'laranja']
    
    # Escolhe randomicamente uma palavra
    palavra = random.choice(palavras)
    
    # Lista  de letras  da palavra
    lista_letras_palavras = [letra for letra in palavra]
    
    # Cria o tabuleiro com o caracter "_" multiplicado pelo comprimento da palavra
    tabuleiro = ["_"] * len(palavra)
    
    # Número de chances
    chances = 6
    
    # Lista para as letras digitadas
    letras_tentativas = []
    
    # Loop enquanto número de chances for maior do que zero
    while chances > 0:
        
        print(display_hangman(chances))
        print("Palavra: ", tabuleiro)
        print("\n")
        
        tentativa = input("\nDigite uma letra: ")
         
        if tentativa in letras_tentativas:
            print("Você já tentou essa letra. Escolha outra!")
            continue    
       
        letras_tentativas.append(tentativa)
        
        if tentativa in lista_letras_palavras:
            
            print("Você acertou a letra!")
            
            for indice in range(len(lista_letras_palavras)):

                # Condicional
                if lista_letras_palavras[indice] == tentativa:
                    tabuleiro[indice] = tentativa
            
            # Se todos os espaços foram preenchidos, o jogo acabou
            if "_" not in tabuleiro:
                print("\nVocê venceu! A palavra era: {}".format(palavra))
                break
        else:
            print("Ops. Essa letra não está na palavra!")
            chances -= 1
    
    if "_" in tabuleiro:
        print("\nVocê perdeu! A palavra era: {}.".format(palavra))
```

```python
if __name__ == "__main__":
    game()
```
